```python
import pandas as pd
import numpy as np
```

```python
#Read CSV and display
df = pd.read_csv('player_data.csv')
df.head(10)
```

```python
#Cheack for null values
df.count()
```

```python
#Clean up null values
df = df.dropna(how='any')
```

```python
#Check datatypes
df.dtypes
```

```python
#Rename columns
df1 = df.rename(columns={'name': 'Player', 'year_start': 'Year Start', 'year_end': 'Year Ended', 'position': 'Position', 'height': 'Height', 'weight': 'Weight', 'birth_date': 'Birth Day', 'college': 'College'})
```

```python
#Import other CSV and display
df2 = pd.read_csv('Seasons_Stats.csv')
df2.head()
```

```python
#Check datatypes
df2.dtypes
```

```python
#Display all columns
columns = df2.columns.tolist()
columns
```

```python
#Remove all irrelevant columns 
df3 = df2.drop(columns=['Unnamed: 0', 'MP', 'FG%', 'Year', 'Age', 'Tm', 'DBPM', '3P', '3PA', '3P%', '2P%', 'GS', 'PER', 'TS%', '3PAr', 'FTr', 'ORB%', 'DRB%', 'TRB%', 'AST%', 'STL%', 'BLK%', 'TOV%', 'USG%', 'blanl', 'OWS', 'DWS', 'WS', 'WS/48', 'blank2', 'OBPM', 'BPM', 'VORP', '2P', '2PA', 'eFG%', 'FT', 'FTA', 'FT%', 'ORB', 'DRB', 'TOV', 'PF'])
```

```python
#Rename Columns
df4 = df3.rename(columns={'Pos': 'Position', 'G': 'Games Played', 'FG': 'Shots Made', 'FGA': 'Shots Attempted', 'TRB': 'Total Rebounds', 'AST': 'Assist', 'STL': 'Steals', 'BLK': 'Blocks', 'PTS': 'Points'})
```

```python
#Change data type
df4[('Games Played')] = df4[('Games Played')].fillna(0.0).astype(int)
df4[('Shots Attempted')] = df4[('Shots Attempted')].fillna(0.0).astype(int)
df4[('Shots Made')] = df4[('Shots Made')].fillna(0.0).astype(int)
df4[('Total Rebounds')] = df4[('Total Rebounds')].fillna(0.0).astype(int)
df4[('Assist')] = df4[('Assist')].fillna(0.0).astype(int)
df4[('Steals')] = df4[('Steals')].fillna(0.0).astype(int)
df4[('Blocks')] = df4[('Blocks')].fillna(0.0).astype(int)
df4[('Points')] = df4[('Points')].fillna(0.0).astype(int)
```

```python
#Strip special characters
df5 = df4.replace('\*','',regex=True)
```

```python
df6 = df5.groupby(['Player']).sum()
```

```python
#Join tables
combind_data = df1.join(df6, on='Player')
```

```python
#Check for null values
combind_data.count()
```

```python
#Remove Null values
cleandata = combind_data.dropna(how='any')
```

```python
#Save cleaned data
cleandata.to_csv('cleandataETL.csv',index= False)
```
